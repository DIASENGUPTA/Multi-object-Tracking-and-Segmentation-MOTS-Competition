```python
import json
import csv
```

```python
with open("metrics.json") as f:
    lines = f.readlines()
    metric=[]
    for line in lines:
        metric.append(json.loads(line))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(metric)
```

```python
df.iloc[5000:].plot.line(y=['total_loss'])
df.iloc[5000:].plot.line(y=['loss_bbox'])
df.iloc[5000:].plot.line(y=['loss_ce'])
df.iloc[5000:].plot.line(y=['loss_mask'])
```

```python
import matplotlib.pyplot as plt

# Import the pandas dataframe


# Get the data from the dataframe
total_loss = df["total_loss"].iloc[5000:]
loss_bbox = df["loss_bbox"].iloc[5000:]
loss_ce = df["loss_ce"].iloc[5000:]
loss_mask = df["loss_mask"].iloc[5000:]

# Plot the data as subplots
fig, axes = plt.subplots(nrows=4, sharex=True, figsize=(10,10))

axes[0].plot(total_loss)
axes[0].set_title("Total Loss")
axes[1].plot(loss_bbox)
axes[1].set_title("Loss Bounding Box")
axes[2].plot(loss_ce)
axes[2].set_title("Loss Cross Entropy")
axes[3].plot(loss_mask)
axes[3].set_title("Loss Mask")

axes[0].plot(total_loss.rolling(window=20).mean(), label="Total Loss Trendline")
axes[1].plot(loss_bbox.rolling(window=20).mean(), label="Loss Bounding Box Trendline")
axes[2].plot(loss_ce.rolling(window=20).mean(), label="Loss Cross Entropy Trendline")
axes[3].plot(loss_mask.rolling(window=20).mean(), label="Loss Mask Trendline")

# Show the plots
plt.show()
```
